#### After looking through the data in the EDA phase, we are moving into the regression analysis part. We will run some regressions to see what variables are explanatory of a winning season in NCAA.

```python
import pandas as pd
pd.options.display.max_columns = None  #This allows us to display all columns in jupyter output
```

```python
Season_Results = pd.read_csv("/Users/ericenglin/Desktop/Data Science Club/DataFiles/RegularSeasonDetailedResults.csv")
Teams = pd.read_csv("/Users/ericenglin/Desktop/Data Science Club/DataFiles/Teams.csv")
```

```python
Season_Results['TeamID']=Season_Results['WTeamID']
Season_Results['CountCol']=1
Season_Results= pd.merge(Season_Results, Teams, on='TeamID')
```

```python
Win = Season_Results.groupby(["TeamName","Season"]).sum()
Win.head()
```

```python
Win['Field_Goal_Percentage']= (Win['WFGM']/Win['WFGA'])
Win['3_Point_Goal_Percentage']= (Win['WFGM3']/Win['WFGA3'])
Win['Free_Throw_Percentage']= (Win['WFTM']/Win['WFTA'])
Win['Point Differential']=(Win['WPF']-Win['LPF'])
Win['Turnover Differential']=(Win['WStl']+Win['WBlk']-Win['WTO'])
Win.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

cols=['Field_Goal_Percentage','3_Point_Goal_Percentage',
      'Free_Throw_Percentage', 'Point Differential', 'Turnover Differential'] 


Win2 = Win[cols]

y=Win['CountCol']
X_train, X_test, y_train, y_test = train_test_split(Win2, y, test_size=0.3)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
import statsmodels.api as sm
ols_model=sm.OLS(y_train, X_train)
result=ols_model.fit()
print(result.summary2())
```

```python
X_train.head()
```

```python
y_pred = result.predict(X_test)
```

```python
y_pred.describe()
```

```python
Win['MSE']= (y_test-y_pred)**2
Win['MSE'].describe()
```
